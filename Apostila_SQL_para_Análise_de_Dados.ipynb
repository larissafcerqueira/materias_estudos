# LIMIT

# VIEWS

```python
'''
Imagine que você deseja fazer uma consulta de todos os intens
da tabela categorias
'''
connection = mysql.connector.connect(
    host=host,
    user=user,
    password=password,
    database=database,
) # estas linhas servem para criar a conexão com o banco de dados
cursor = connection.cursor() #criando o cursor para apontar para o banco de dados
query = '''
SELECT *
FROM categorias;
'''
cursor.execute(query) # Execução da consulta
result = cursor.fetchall() # armazenando os resultados na variável results
cast = pd.DataFrame(result,columns=cursor.column_names)
display(cast) # exibindo a tabela pandas
cursor.close() # fechando o cursor
connection.close() # fechando a conexão
```

```python
'''
Imagine que você deseja gravar o texto da consulta anterior no banco de dados
'''
# Criando uma VIEW
connection = mysql.connector.connect(
    host=host,
    user=user,
    password=password,
    database=database,
) # estas linhas servem para criar a conexão com o banco de dados
cursor = connection.cursor() #criando o cursor para apontar para o banco de dados
query = '''
CREATE OR REPLACE VIEW exemplo AS
SELECT *
FROM categorias;
'''
cursor.execute(query) # Execução da consulta
result = cursor.fetchall() # armazenando os resultados na variável results
cast = pd.DataFrame(result,columns=cursor.column_names)
display(cast) # exibindo a tabela pandas
cursor.close() # fechando o cursor
connection.close() # fechando a conexão
```

```python
# Executando uma VIEW
connection = mysql.connector.connect(
    host=host,
    user=user,
    password=password,
    database=database,
) # estas linhas servem para criar a conexão com o banco de dados
cursor = connection.cursor() #criando o cursor para apontar para o banco de dados
query = '''
SELECT *
FROM exemplo;
'''
cursor.execute(query) # Execução da consulta
result = cursor.fetchall() # armazenando os resultados na variável results
cast = pd.DataFrame(result,columns=cursor.column_names)
display(cast) # exibindo a tabela pandas
cursor.close() # fechando o cursor
connection.close() # fechando a conexão
```

```python
# Excluindo uma VIEW
connection = mysql.connector.connect(
    host=host,
    user=user,
    password=password,
    database=database,
) # estas linhas servem para criar a conexão com o banco de dados
cursor = connection.cursor() #criando o cursor para apontar para o banco de dados
query = '''
DROP VIEW
exemplo;
'''
cursor.execute(query) # Execução da consulta
result = cursor.fetchall() # armazenando os resultados na variável results
cast = pd.DataFrame(result,columns=cursor.column_names)
display(cast) # exibindo a tabela pandas
cursor.close() # fechando o cursor
connection.close() # fechando a conexão
```
