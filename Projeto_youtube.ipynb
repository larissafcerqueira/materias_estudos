```python
import pandas as pd
```

```python
df = pd.read_csv('https://storage.googleapis.com/gabarito-projeto/dados-brutos/base_do_projeto%20-%20Vendas.csv')
display(df)
```

```python
df.info()
```

```python
df.rename(columns={'SKU':'id','Produto':'produto','Qtd Vendida':'quantidade',
                   'Data':'data','Loja':'loja','Preço Unitário':'preco',
                   'Nome':'nome','Sobrenome':'sobrenome','Gênero':'genero'},inplace=True)
df.head(1)
```

```python
df.drop('Unnamed: 9',axis=1,inplace=True)
df.head(1)
```

```python
df['comissao'] = (df['preco']*df['quantidade']) * 0.03
df.head(5)
```

```python
df.to_csv('vendas.csv',index=False)
```
