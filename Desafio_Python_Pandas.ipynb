```python

import pandas as pd
import numpy as np

# Carregar a base de dados
df = pd.read_csv('saude_do_sono_estilo_vida.csv')


# 1
df.rename(columns={
    'ID': 'Identificador',
    'Pressão sanguíneaaaa': 'Pressão sanguínea',
    'Ocupação': 'Profissão',
    'Categoria BMI': 'Categoria IMC'
}, inplace=True)

print(""" Ao visualizar a base você percebeu que seria melhor alterar o nome de
algumas colunas. Mude o ‘ID’ para ‘Identificador’, corrija o nome da
coluna que indica a pressão sanguínea, mude a coluna ‘Ocupação’ para
‘Profissão’, a coluna ‘Categoria BMI’ está em parte em inglês, substitua
para ‘Categoria IMC’.""")
print(df.head())


#2
sleep_stats = df.groupby('Profissão')['Duração do sono'].agg(['mean', 'median', lambda x: x.mode()[0]])
sleep_stats.columns = ['Média', 'Mediana', 'Moda']

print("""\n2. Qual é a média, a moda e a mediana de horas de sono para cada uma
das profissões?""")
print(sleep_stats)



#3
software_engineers = df[df['Profissão'] == 'Eng. de Software']
percent_obese_software_engineers = (software_engineers['Categoria IMC'] == 'Obesidade').mean() * 100

print("""\n3. Das pessoas que atuam com engenharia de software qual a
porcentagem de obesos?""")
print(f"{percent_obese_software_engineers:.2f}%")



#4
profissoes = df[df['Profissão'].isin(['Advogado(a)', 'Representante de Vendas'])]
mean_sleep_profissoes = profissoes.groupby('Profissão')['Duração do sono'].mean()

print("""\n4. De acordo com os dados, advogar ou ser representante de vendas faz
você dormir menos?""")
print(mean_sleep_profissoes)


#5
enfermagem_medicina = df[df['Profissão'].isin(['Enfermeiro(a)', 'Médico(a)'])]
mean_sleep_enfermagem_medicina = enfermagem_medicina.groupby('Profissão')['Duração do sono'].mean()

print("""\n5. Entre quem fez enfermagem e quem fez medicina, quem tem menos
horas de sono?""")
print(mean_sleep_enfermagem_medicina)


#6
subset = df[['Identificador', 'Gênero', 'Idade', 'Pressão sanguínea', 'Frequência cardíaca']]

print("""\n6.  Faça um subconjunto com as colunas Identificador, Gênero, Idade,
Pressão sanguínea e Frequência cardíaca.""")
print(subset.head())


#7
least_frequent_profession = df['Profissão'].value_counts().idxmin()

print("""\n7. Descubra qual a profissão menos frequente no conjunto.""")
print(least_frequent_profession)


#8
df['Pressão sistólica'] = df['Pressão sanguínea'].apply(lambda x: int(x.split('/')[0]))
mean_blood_pressure_gender = df.groupby('Gênero')['Pressão sistólica'].mean()

print("""\n8. Quem tem maior pressão sanguínea média, homens ou mulheres?""")
print(mean_blood_pressure_gender)


#9
mode_sleep_duration = df['Duração do sono'].mode()[0]
predominant_8_hours = mode_sleep_duration == 8

print("""\n9. É predominante entre os participantes dormir 8 horas por dia""")
print(predominant_8_hours)


#10
high_heart_rate = df[df['Frequência cardíaca'] > 70]['Passos diários'].mean()
low_heart_rate = df[df['Frequência cardíaca'] <= 70]['Passos diários'].mean()

print("""\n10. Pessoas com frequências cardíacas acima de 70 dão mais passos que
pessoas com frequência cardíaca menor ou igual a 70?""")
print(f"Frequência cardíaca > 70: {high_heart_rate}")
print(f"Frequência cardíaca <= 70: {low_heart_rate}")


```
